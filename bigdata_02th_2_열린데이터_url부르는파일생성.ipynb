```python
%%writefile src/ds_open_subwayTime.py

import os
import requests
import json
from pymongo import MongoClient
import mylib

Client= MongoClient()
_db=Client['ds_open_subwayPassengerDb']
_table=_db['db_open_subwayTable']

def saveJson(_fname,_data):
    import io
    with io.open(_fname, 'a', encoding='utf8') as json_file:
        _j= json.dump(_data, json_file, ensure_ascii=False)
        json_file.write(str(_j)+'\n')
        
def readJson(_fname):
    for line in open(_fname, 'r').readlines():
        _j=json.load(line)
        print(_j['row'])
        
def saveDB(_data):
    _table.insert_one(_data)
    
def readDB():
    for tweet in _table.find():
        print(tweet['CardSubwayTime'], tweet['row'])
        
def saveFile(_fname, _data):
    fp= open(_fname, 'a')
    fp.write(_data+'\n')
    
def doIt():
    _key= mylib.getKey('src/key.properties')
    KEY=_key['dataseoul']
    URL= 'http://openapi.seoul.go.kr:8088'
    TYPE= 'json'
    SERVICE= 'CardSubwayTime'
    START_INDEX=1
    END_INDEX=10
    USE_MON='202005'
    maxIter=10
    _iter=0
    _jfname='src/ds_open_subwayTime.json'
    
    while _iter<maxIter:
        _url="/".join([URL, KEY,TYPE,SERVICE,str(START_INDEX),str(END_INDEX),USE_MON])
        print(_url+"\n")
        rp=requests.get(_url)
        _json=rp.json()
        print(_json)
        saveJson(_jfname,_json)
        saveDB(_json)
        
        START_INDEX+=10
        END_INDEX+=10
        _iter+=1
        
if __name__=='__main__':
    doIt()
```
