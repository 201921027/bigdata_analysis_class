```python
import urllib.parse

_type= 'json'
_service= 'VwsmTrdarWrcPopltnQq'
_start_index= 1
_end_index= 100
_year='2020'

_url= 'http://openAPI.seoul.go.kr:8088/'
#url= os.path.join(_key, _type, _service, str(_start_index), str(_end_index), _year)
url= "/".join([_key, _type, _service, str(_start_index), str(_end_index), _year])
url= urllib.parse.urljoin(_url, url)
print(url)
```

```python
import requests
import json

_fname='src/file_3th.json'
fp= open(_fname, 'w')
count=0
_start_index=0
_end_index=100

while True:
    rp= requests.get(url)
    trade= rp.json()
    print_r=trade["VwsmTrdarWrcPopltnQq"]["row"]
    if count==0:
        print("------------ START  ------------")
        total_count= trade['VwsmTrdarWrcPopltnQq']['list_total_count']
        print('total_count: ', total_count)
        print("first data: ", end=' ')
        print(print_r)
        count+=1

    fp.write(str(trade)+"\n")
    _start_index+=100
    _end_index+=100
    url= "/".join([_key, _type, _service, str(_start_index), str(_end_index), _year])
    url= urllib.parse.urljoin(_url, url)
    if _end_index> total_count:
        print("last data: ", end=' ')
        print(print_r)
        print("# of data: ", (_end_index)-100)
        print("------------ FINISH ------------")
        break

fp.close()
```
