```python
%%writefile src/ds_open_subway_iter_json.py

import requests
import os
import mylib

def doit():
    keyPath= os.path.join(os.getcwd(), 'src', 'key.properties')
    key=mylib.getKey(keyPath)
    KEY= str(key['dataseoul'])
    TYPE='json'
    SERVICE='SearchSTNBySubwayLineInfo'
    START_INDEX=str(1)
    END_INDEX=str(10)
    LINE_NUM=str(2) #2호선
    
    startIndex=1
    endIndex=10
    list_total_count=0
    
    while True:
        #url 만들기
        START_INDEX=str(startIndex)
        END_INDEX=str(endIndex)
        params= "/".join([KEY,TYPE,SERVICE,START_INDEX,END_INDEX,'','',LINE_NUM])
        #print(params)
        #print(params[30:])
        _url= 'http://openapi.seoul.go.kr:8088'
        url="/".join([_url, params])
        #print(url)
    
        #url 호출
        r= requests.get(url) # r에 url의 response가 삽입됨
        stations=r.json() # json 파싱
        if(startIndex==1):
            list_total_count= stations['SearchSTNBySubwayLineInfo']['list_total_count']
            print("- Total Count: ", list_total_count)
        for e in stations['SearchSTNBySubwayLineInfo']['row']:
            print("{0:4s}\t{1:15s}\t{2:3s}\t{3:2s}".format(e['STATION_CD'], e['STATION_NM'],e['LINE_NUM'],e['FR_CODE']))
        
        startIndex+=10
        endIndex+=10
        if(endIndex>list_total_count):
            #print("----- Ending endIndex=", endIndex)
            #break
            num=0
            
            START_INDEX=str(startIndex)
            END_INDEX=str(list_total_count)
            params= "/".join([KEY,TYPE,SERVICE,START_INDEX,END_INDEX,'','',LINE_NUM])
            #print(params)
            #print(params[30:])
            _url= 'http://openapi.seoul.go.kr:8088'
            url="/".join([_url, params])
            #print(url)
    
            #url 호출
            r= requests.get(url) # r에 url의 response가 삽입됨
            stations=r.json() # json 파싱
            
            for e in stations['SearchSTNBySubwayLineInfo']['row']:
                num+=1
                print("{0:4s}\t{1:15s}\t{2:3s}\t{3:2s}".format(e['STATION_CD'], e['STATION_NM'],e['LINE_NUM'],e['FR_CODE']))
            print("----- Ending endIndex=", endIndex-10+num)
            break

if __name__=="__main__":
    doit()
```

```python
%%writefile src/ds_open_subway_xml.py

import lxml
import lxml.etree
import requests
import os
import mylib

def doit():
    keyPath= os.path.join(os.getcwd(), 'src', 'key.properties')
    key=mylib.getKey(keyPath)
    KEY= str(key['dataseoul'])
    TYPE='xml'
    SERVICE='SearchSTNBySubwayLineInfo'
    START_INDEX=str(1)
    END_INDEX=str(10)
    LINE_NUM=str(2) #2호선
    
    startIndex=1
    endIndex=10
    list_total_count=0
    
    while True:
        #url 만들기
        START_INDEX=str(startIndex)
        END_INDEX=str(endIndex)
        params= "/".join([KEY,TYPE,SERVICE,START_INDEX,END_INDEX,'','',LINE_NUM])
        #print(params)
        #print(params[30:])
        _url= 'http://openapi.seoul.go.kr:8088'
        url="/".join([_url, params])
        #print(url)
    
        #url 호출
        data= requests.get(url).text # r에 url의 response가 삽입됨
        tree=lxml.etree.fromstring(data.encode('utf-8'))
        
        if START_INDEX=='1':
            for node in tree.xpath('//list_total_count'):
                list_total_count=int(node.text)
                print("-Total count= ", list_total_count)
            
        for node in tree.xpath('//STATION_NM'):
            print(node.text, end=", ")
        
        startIndex+=10
        endIndex+=10
        if(endIndex>list_total_count):
            #print("----- Ending endIndex=", endIndex)
            #break
            
            START_INDEX=str(startIndex)
            END_INDEX=str(list_total_count)
            params= "/".join([KEY,TYPE,SERVICE,START_INDEX,END_INDEX,'','',LINE_NUM])
            #print(params)
            #print(params[30:])
            _url= 'http://openapi.seoul.go.kr:8088'
            url="/".join([_url, params])
    
            #url 호출
            data= requests.get(url).text
            tree=lxml.etree.fromstring(data.encode('utf-8'))
            
            for node in tree.xpath('//STATION_NM'):
                print(node.text, end=", ")
            print("\n----- Ending endIndex=", list_total_count)
            break

if __name__=="__main__":
    doit()
```
